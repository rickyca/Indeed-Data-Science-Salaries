```python
def title_roman_numbers_replacer(ds):
    '''
    Replaces roman numbers in column title of a dataframe
    '''
    ds['title'] = ds['title'].str.replace(' I ',' 1 ')
    ds['title'] = ds['title'].str.replace(' II ',' 2 ')
    ds['title'] = ds['title'].str.replace(' III ',' 3 ')
    ds['title'] = ds['title'].str.replace(' IV ',' 4 ')
    return ds

# Replace roman numbers
ds = title_roman_numbers_replacer(ds)
```

```python
# Set up 
y = ds['salary']
X = ds.drop('salary', axis=1)
```

```python
# Train Test split (without salary as column)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Vectorize
vectorizer = CountVectorizer(min_df=5, ngram_range=(2,4), max_features=10000, \
                             stop_words='english')

# Fit and transform test data
word_matrix = vectorizer.fit_transform(X_train['title'])
# Transform test data
word_matrix_test = vectorizer.transform(X_test['title'])
```

```python
# Make a data frame of the tokenized data
word_df = pd.DataFrame(word_matrix.toarray(), columns=vectorizer.get_feature_names())
word_df_test = pd.DataFrame(word_matrix_test.toarray(), columns=vectorizer.get_feature_names())
```

```python
# Plot N Grams count
plt.figure(figsize=(14,16))
word_df.sum().sort_values().plot(kind='barh')
plt.show()
```

```python
# Unsupervized learning
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ys_train = ss.fit_transform(y_train.values.reshape(-1,1))
```

```python
full_t = Xv_train.copy()
full_t['salary'] = ys_train
full_t.shape
```

```python
ss2 = StandardScaler()
full_t['location'] = ss2.fit_transform(full_t['location'])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# KMeans Scorer function
def kmeans_scorer(X, n_clusters=[2,10]):
    inertia = []
    silhouette = []
    for k in range(n_clusters[0], n_clusters[1]):
        km = KMeans(n_clusters=k)
        y_pred = km.fit_predict(X)
        s = silhouette_score(X, y_pred)
        i = km.inertia_
        silhouette.append(s)
        inertia.append(i)
    plt.plot(range(n_clusters[0], n_clusters[1]), inertia/max(inertia))
    plt.plot(range(n_clusters[0], n_clusters[1]), silhouette)
```

```python
kmeans_scorer(full_t)
plt.show()
```

```python
# Using the elbow criteria I pick 4 clusters
km = KMeans(n_clusters=4)
full_yt = km.fit_predict(full_t)
```
